```python
date_columns = [
    "PQ First Sent to Client Date",
    "PO Sent to Vendor Date",
    "Scheduled Delivery Date",
    "Delivered to Client Date",
    "Delivery Recorded Date"
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

```python
df["Scheduled Delivery Date"].head(10)
```

```python
df["Weight (Kilograms)"] = pd.to_numeric(
    df["Weight (Kilograms)"], errors="coerce"
)

df["Freight Cost (USD)"] = pd.to_numeric(
    df["Freight Cost (USD)"], errors="coerce"
)
```

```python
df.isnull().sum()
```

```python
df["Shipment Mode"] = df["Shipment Mode"].fillna(df["Shipment Mode"].mode()[0])
```

```python
df["Dosage"] = df["Dosage"].fillna("Unknown")
```

```python
df["Weight (Kilograms)"] = df["Weight (Kilograms)"].fillna(df["Weight (Kilograms)"].median())

df["Freight Cost (USD)"] = df["Freight Cost (USD)"].fillna(df["Freight Cost (USD)"].median())

df["Line Item Insurance (USD)"] = df["Line Item Insurance (USD)"].fillna(df["Line Item Insurance (USD)"].median())
```

```python
df.isnull().sum()
```

```python
df["Shipment Mode"].value_counts()
```

```python
df["Country"].value_counts()
```

```python
df["Delivery Delay"] = (
    df["Delivered to Client Date"] -
    df["Scheduled Delivery Date"]
).dt.days
```

```python
df[["Scheduled Delivery Date",
    "Delivered to Client Date",
    "Delivery Delay"]].head()
```

```python
df["Delivery Delay"].mean()
```

```python
delayed = df[df["Delivery Delay"] > 0]

delayed.head()
```

```python
country_shipments = df.groupby("Country")["ID"].count().sort_values(ascending=False)

country_shipments
```

```python
vendor_cost = df.groupby("Vendor")["Line Item Value"].sum().sort_values(ascending=False)

vendor_cost
```

```python
product_value = df.groupby("Product Group")["Line Item Value"].sum().sort_values(ascending=False)

product_value
```

```python
shipment_delay = df.groupby("Shipment Mode")["Delivery Delay"].mean()

shipment_delay
```

```python
country_shipments.head(10).plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Countries by Shipments")
plt.xlabel("Country")
plt.ylabel("Number of Shipments")
plt.xticks(rotation=45)
plt.show()
```

```python
shipment_mode = df["Shipment Mode"].value_counts()

shipment_mode.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Shipment Mode Distribution")
plt.ylabel("")
plt.show()
```

```python
delivery_trend = df.groupby(
    df["Scheduled Delivery Date"].dt.to_period("M")
)["ID"].count()

delivery_trend.index = delivery_trend.index.astype(str)

delivery_trend.plot(kind="line", figsize=(12,5))

plt.title("Delivery Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Deliveries")
plt.xticks(rotation=45)
plt.show()
```
